```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset.describe()
```

```python
diabetes_dataset.info()
```

```python
diabetes_dataset['Outcome'].value_counts()  ## as outcome is unbalanced, it can cause biased model
```

## separating the features and target

```python
features = diabetes_dataset.drop(columns='Outcome',axis=1)
features.head()
```

```python
target = diabetes_dataset['Outcome']
target.head()
```

## Standardise values to a similar range.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

#### Below fit and transform can be done in one step using fit and transform

```python
scaler.fit(features)
```

```python
standardized_data = scaler.transform(features)
```

```python
standardized_data
```

```python
features = standardized_data
```

## Split data into training and testing data set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=2)
```

```python
x_test
```

```python
print(features.shape, x_train.shape, x_test.shape)
```

## Training the model

```python
classifier = LogisticRegression(random_state=0, max_iter=1000)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_prediction = classifier.predict(x_test)
```

## Check for the accuracy of generated model

```python
from sklearn.metrics import accuracy_score
```

```python
test_accuracy = accuracy_score(y_test,y_prediction)
```

```python
print(test_accuracy)
```

## Testing the model

```python
input_data = (5,166,72,19,175,25.8,0.58,51)
input_data
```

```python
# converting tuple to np array format
input_data_arrayformat = np.asarray(input_data)
input_data_arrayformat
```

```python
# Reshape the data
input_data_reshaped = input_data_arrayformat.reshape(1,-1)
input_data_reshaped
```

```python
#transform the data
std_data = scaler.transform(input_data_reshaped)
```

```python
std_data
```

```python
# predict the outcome
prediction = classifier.predict(std_data)

if (prediction==0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
