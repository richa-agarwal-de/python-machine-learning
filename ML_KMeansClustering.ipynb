# K-Means Clustering - Unsupervised Learning

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('income1.csv')
df.head()
```

```python
df.count()
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
print(y_predicted)
```

```python
df['cluster'] = y_predicted
df.head(25)
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='blue')
plt.scatter(df3.Age,df3['Income($)'],color='red')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
## Preprocessing using MinMazScaler

df.head()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Income($)']])
```

```python
df['Income($)'] = scaler.transform(df[['Income($)']])
df.head()
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])

df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
print(y_predicted)
```

```python
km.cluster_centers_
```

```python
df['cluster'] = y_predicted
df.head(25)
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='blue')
plt.scatter(df3.Age,df3['Income($)'],color='red')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
km.cluster_centers_
```

```python
## Elbow method to figure out value of 'K'

sse = []

k_rng = range(1,10)
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('SSE')

plt.plot(k_rng,sse)
```

# K-Means++ clustering

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt
```

```python
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Math':    [88, 78, 95, 60, 55, 62, 80, 90, 58, 85],
    'Science': [92, 85, 96, 65, 58, 60, 83, 91, 59, 87]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
X = df[['Math','Science']] ## removing student name and keeping only the marks columns
print(X)
```

```python
## Apply KMeans++ clustering Algorithm

model = KMeans(n_clusters=2,init = 'k-means++',random_state = 42)

df['Cluster'] = model.fit_predict(X)
print(df)
```

```python
model.cluster_centers_
```

```python
plt.scatter(df['Math'],df['Science'],c=df['Cluster'],cmap='viridis',s=200)
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c='red',marker='X',s=200)
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.title('K-Means++ Clustering of Students')
plt.grid(True)
plt.show()
```

```python
print('KMeans++ iterations',model.n_iter_)
```

```python
## Apply KMeans clustering Algorithm for comparison

model1 = KMeans(n_clusters=2,random_state = 42)

df['Cluster1'] = model1.fit_predict(X)
print(df)
```

```python
model1.cluster_centers_
```

```python
plt.scatter(df['Math'],df['Science'],c=df['Cluster1'],cmap='viridis',s=200)
plt.scatter(model1.cluster_centers_[:,0],model1.cluster_centers_[:,1],c='red',marker='X',s=200)
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.title('K-Means Clustering of Students')
plt.grid(True)
plt.show()
```

```python
print('KMeans iterations',model1.n_iter_)
```
