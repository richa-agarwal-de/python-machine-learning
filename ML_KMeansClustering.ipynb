```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('income1.csv')
df.head()
```

```python
df.count()
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
print(y_predicted)
```

```python
df['cluster'] = y_predicted
df.head(25)
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='blue')
plt.scatter(df3.Age,df3['Income($)'],color='red')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
## Preprocessing using MinMazScaler

df.head()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Income($)']])
```

```python
df['Income($)'] = scaler.transform(df[['Income($)']])
df.head()
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])

df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
print(y_predicted)
```

```python
km.cluster_centers_
```

```python
df['cluster'] = y_predicted
df.head(25)
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='blue')
plt.scatter(df3.Age,df3['Income($)'],color='red')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
km.cluster_centers_
```

```python
## Elbow method to figure out value of 'K'

sse = []

k_rng = range(1,10)
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('SSE')

plt.plot(k_rng,sse)
```
