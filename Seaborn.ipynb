## Scatter plot

```python
sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm'
)
```

```python
sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'sex'
)
```

```python
sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'island'
)
```

```python
sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'species'
)
```

```python
## specify color palette

island_colors = {'Torgersen':'green',
                'Biscoe':'blue',
                'Dream':'orange'}

sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'island',
    palette = island_colors
)
```

```python
## set style function  darkgrid, whitegrid, dark, white, ticks.
## by default it is ticks

sns.set_style('dark')

sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'island',
    palette = island_colors
)

```

```python
## set context: paper, notebok, talk, poster

sns.set_context('paper')

sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'island',
    palette = island_colors
)

```

```python
sns.set_context('talk')
sns.set_style('white')

sns.scatterplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'island',
    palette = island_colors
)
```

```python
## scatter plot for categorical value

sns.set_context('notebook')

sns.scatterplot(
    data = penguins,
    x = 'species',
    y = 'body_mass_g',
    hue = 'island',
    palette = island_colors
)
```

## Strip plot

```python
sns.stripplot(
    data = penguins,
    x = 'species',
    y = 'body_mass_g',
    hue = 'island',
    palette = island_colors
)
```

```python
sns.stripplot(
    data = penguins,
    x = 'species',
    y = 'body_mass_g',
    hue = 'island',
    palette = island_colors,
    dodge = True
)
```

## Swarm Plots

```python
sns.swarmplot(
    data = penguins,
    x = 'species',
    y = 'body_mass_g',
    hue = 'island'
)
```

## Histogram

```python
sns.set_context('notebook')

sns.histplot(
    data = penguins,
    x = 'body_mass_g'
)
```

```python
sns.histplot(
    data = penguins,
    x = 'body_mass_g',
    hue = 'sex'
)
```

## Regression

```python
sns.regplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm'
)
```

```python
sns.regplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    color = 'green'
)
```

## Line plots

```python
sns.set_style('whitegrid')

sns.lineplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm'
)
```

```python
sns.lineplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'sex'
)
```

## Joint plot

```python
sns.jointplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm'
)
```

```python
sns.jointplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'sex'
)
```

```python
sns.jointplot(
    data = penguins,
    x = 'body_mass_g',
    y = 'flipper_length_mm',
    hue = 'sex',
    kind = 'kde'
)
```

## Bar plot

```python
sns.barplot(
    data = penguins,
    x = 'species',
    y = 'body_mass_g',
    hue = 'sex',
    palette = ['skyblue','pink']

)
```
